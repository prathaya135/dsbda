```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("iris.csv")
df
```

```python
 sns.countplot(x='petal_width',data=df);
```

```python
sns.countplot(x='species', data=df)
plt.xlabel("species (0 = sepal_length, 1= petal_length)")
```

```python
pd.crosstab(df.sepal_length,df.species).plot(kind="bar",figsize=(20,6))
plt.title('iris frequency')
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
plt.show()
```

```python
sns.boxplot(data=df, x='sepal_length')
```

```python
sns.histplot(data=df, x='sepal_length')
```

```python
sns.lineplot(data=df, x='sepal_length',y='petal_length')
```

```python
sns.lineplot(data=df, x='sepal_length',y='petal_length',hue='species')
```

```python


# Create a scatter plot
sns.scatterplot(data=df, x="sepal_length", y="sepal_width", hue="species")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Length vs Sepal Width")
plt.legend(title="Species")
plt.show()
```
